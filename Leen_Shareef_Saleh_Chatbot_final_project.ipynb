```python
import datetime

print("hello, welcome to hotel chatbot")
print("type 'hi' or 'hello' to start a conversation. Type 'quit' to exit")

while True:
    user_input = input("> ").lower()

    if user_input in ["hi", "hello"]:
        print("Hello how can i assist you today?")
        print("1. book a room")
        print("2. get information")
        print("3. exit")

    elif user_input == "1":
        print("Let's start with the booking")
        while True:
            print("When would you like to book your stay? Please enter a date in this form please (2024-06-15):")
            date = input("> ")

            # Convert user input date string to datetime object
            user_date = datetime.datetime.strptime(date, "%Y-%m-%d").date()
            current_date = datetime.date.today()

            # Check if user input date is before the current local date
            if user_date < current_date:
                print("Invalid date. Please enter a date on or after today.")
            else:
                break

        print(f"Booking date: {date}")

        print("What type of room would you like to book?")
        print("1. single")
        print("2. double")
        print("3. suite")
        room_type = input("> ")
        if room_type == "1":
            print("you have selected: single room")
        elif room_type == "2":
            print("you have selected: double room")
        elif room_type == "3":
            print("you have selected: Suite")

        print("What type of meal plan do you prefer?")
        print("1. Room only")
        print("2. Breakfast")
        print("3. all inclusive")
        meal_plan = input("> ")
        if meal_plan == "1":
            print("you selected: Room only")
        elif meal_plan == "2":
            print("you selected: Breakfast")
        elif meal_plan == "3":
            print("you selected: All Inclusive")

        print("Do you need a car parking space?")
        print("1. yes")
        print("2. no")
        parking = input("> ")
        if parking == "1":
            print("You selected: yes")
        elif parking == "2":
            print("You selected: no")

        print("Do you have any special requests?")
        print("1. yes")
        print("2. no")
        special_requests = input("> ")
        if special_requests == "1":
            print("Please enter your special requests:")
            requests = input("> ")
            print(f"Special requests: {requests}")
        elif special_requests == "2":
            print("No special requests")

        print("thank you for your responses. we will process your booking request and get back to you shortly. Type 'quit' to exit")
        print("If you need more details or have any questions, feel free to call the reception at +123456789.")

    elif user_input == "2":
        print("Sure! If you need any information or assistance, please call our reception at +123456789. Type 'quit' to exit.")

    elif user_input == "3" or user_input == "quit":
        print("Goodbye!")
        break

    else:
        print("I didn't understand that. Please type 'hi' or 'hello' to start or 'quit' to exit.")
```
